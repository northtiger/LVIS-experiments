```julia
using MLDataPattern
using MXNet
using Plots
gr()
```

```julia
include("learning.jl")
```

```julia
m = randn()
b = randn()
f = x -> m * x + b
X = rand(1000)
y = f.(X) .+ 0.01 .* randn(size(X))
batch_size = 1
(X_train_validate, y_train_validate), (X_test, y_test) = splitobs((X, y), at=0.75)
(X_train, y_train), (X_validate, y_validate) = splitobs((X_train_validate, y_train_validate), at=0.5)

train_provider = mx.ArrayDataProvider(:x=>collect(X_train), :u_label=>collect(y_train), batch_size=batch_size)
eval_provider = mx.ArrayDataProvider(:x=>collect(X_validate), :u_label=>collect(y_validate), batch_size=batch_size)
test_provider = mx.ArrayDataProvider(:x=>collect(X_test), :u_label=>collect(y_test), batch_size=batch_size)
```

```julia
mlp = @mx.chain mx.Variable(:x) => 
mx.FullyConnected(name=:fc1, num_hidden=1) =>
mx.LinearRegressionOutput(name=:u)

model = mx.FeedForward(mlp, context=mx.cpu())
optimizer = mx.SGD(lr=0.1, momentum=0.9)

mx.fit(model, optimizer, train_provider, n_epoch=10, verbosity=0)
```

```julia
y_predict = vec(mx.predict(model, test_provider))
plt = plot(X_test[1:100], y_test[1:100], line=nothing, markersize=5, marker=:circle)
plot!(plt, X_test[1:10], y_predict[1:10])
```
