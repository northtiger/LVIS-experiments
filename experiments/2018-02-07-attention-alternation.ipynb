```julia
using Revise
```

```julia
using Flux
using Plots; gr()
using ProgressMeter
```

```julia
import FluxExtensions
```

```julia
function oracle(x)
    if x < 2
        -0.5x + 3
    else
        2x - 7
    end
end
```

```julia
xx = linspace(0, 5, 100)
plot(xx, oracle.(xx))

train_x = linspace(0, 5, 10)
train_y = oracle.(train_x)

wrap(x) = [x]
unwrap(x) = (@assert length(x) == 1; Flux.Tracker.value(first(x)))

train_data = collect(zip(wrap.(train_x), wrap.(train_y)))
```

```julia
signals = Dense(1, 2)
weights = Chain(Dense(1, 4, leakyrelu), Dense(4, 2, leakyrelu), softmax)
model = FluxExtensions.Attention(signals, weights)

loss = (x, y) -> Flux.mse(model(x), y)
opt = Flux.Optimise.ADADelta(params(model))

signals_opt = Flux.Optimise.ADADelta(params(model.signals))
weights_opt = Flux.Optimise.ADADelta(params(model.weights))
```

```julia
xx = linspace(0, 5, 100)
plt = plot(xx, oracle.(xx))
plot!(plt, xx, unwrap.(model.(wrap.(xx))))
scatter!(plt, train_x, train_y)
```

```julia
@showprogress for i in 1:100
    for i in 1:100
        Flux.train!(loss, train_data, signals_opt)
    end
    for i in 1:100
        Flux.train!(loss, train_data, weights_opt)
    end
end
```

```julia
xx = linspace(0, 5, 100)
plt = plot(xx, oracle.(xx))
plot!(plt, xx, unwrap.(model.(wrap.(xx))))
scatter!(plt, train_x, train_y)
```

```julia
@showprogress for i in 1:10000
    Flux.train!(loss, train_data, opt)
end
```
