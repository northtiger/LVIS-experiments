```julia
using Revise
```

```julia
using MeshCat
using MeshCatMechanisms
using RigidBodyDynamics
using Gurobi
using LCPSim
using LearningMPC
using LearningMPC.Models
```

```julia
robot = CartPole()
mvis = MechanismVisualizer(robot)
IJuliaCell(mvis)
```

```julia
params = MPCParams(robot)
lqrsol = LQRSolution(robot, params)
x0 = nominal_state(robot)
velocity(x0)[1] += 3
results = LCPSim.simulate(x0, lqrsol, environment(robot), params.Δt, 200, GurobiSolver(Gurobi.Env(), OutputFlag=0));
```

```julia
for r in results
    set_configuration!(mvis, configuration(r.state))
    sleep(params.Δt)
end
```
