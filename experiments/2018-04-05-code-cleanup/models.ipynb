```julia
using Revise
```

```julia
using MeshCatMechanisms
using LearningMPC
using LearningMPC.Models
using MeshCat
using Gurobi
using LCPSim
using RigidBodyDynamics
```

```julia
robot = Hopper()
mvis = MechanismVisualizer(robot)
IJuliaCell(mvis)
```

```julia
params = MPCParams(robot)
lqrsol = LQRSolution(robot, params)
mpc_controller = LearningMPC.MPCController(robot, params, lqrsol, [lqrsol])
results = LCPSim.simulate(nominal_state(robot), mpc_controller, robot.environment, params.Δt, 100, GurobiSolver(Gurobi.Env(), OutputFlag=0));
```

```julia
LearningMPC.playback(mvis, results, params.Δt)
```

```julia
robot = CartPole()
mvis = MechanismVisualizer(robot)
IJuliaCell(mvis)
```

```julia
params = MPCParams(robot)
lqrsol = LQRSolution(robot, params)
mpc_controller = LearningMPC.MPCController(robot, params, lqrsol, [lqrsol])
x0 = nominal_state(robot)
set_velocity!(x0, findjoint(mechanism(robot), "cart_to_pole"), [1.0])
results = LCPSim.simulate(x0, mpc_controller, robot.environment, params.Δt, 150, GurobiSolver(Gurobi.Env(), OutputFlag=0));
```

```julia
LearningMPC.playback(mvis, results, params.Δt)
```
